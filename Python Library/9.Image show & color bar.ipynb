```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
img=mpimg.imread("pie_chart.png")
img
```

```python
#To show the image
plt.imshow(img)
plt.show()
```

```python
#To remove axis we get
plt.axis("off")
plt.imshow(img)
plt.show()
```

```python
#To increase figure size
plt.figure(figsize=(16,9))
plt.axis("off")
plt.imshow(img)
plt.show()
```

```python
#To show the color bar
plt.figure(figsize=(16,9))
plt.axis("off")
plt.imshow(img)
plt.colorbar()
plt.show()
```

```python
#To see the color bar clearly we get
plt.figure(figsize=(16,9))
plt.axis("off")
plt.imshow(img,cmap="Accent")
plt.colorbar()
plt.show()
```

```python
#To change color of each bounded region
single_channel=img[:,:,2]
plt.figure(figsize=(16,9))
plt.axis("off")
plt.imshow(img)
plt.show()
```
