```python
import matplotlib.pyplot as plt
classes=["python","R","ML","AI","DS"]
x=[45,15,35,25,30]
plt.pie(x)
plt.show()
```

```python
classes=["python","R","ML","AI","DS"]
x=[45,15,35,25,30]
plt.pie(x,labels=classes)
plt.show()
```

```python
#To slice the chart
classes=["python","R","ML","AI","DS"]
x=[45,15,35,25,30]
explode=[0,0,0.3,0,0]
plt.pie(x,labels=classes,explode=explode)
plt.show()
```

```python
#To provide each section with diffrent color
classes=["python","R","ML","AI","DS"]
x=[45,15,35,25,30]
explode=[0,0,0.3,0,0]
colors=["c","b","r","y","g"]
plt.pie(x,labels=classes,explode=explode,colors=colors)
plt.show()
```

```python
#To Regeon bounded with percentage
classes=["python","R","ML","AI","DS"]
x=[45,15,35,25,30]
explode=[0,0,0.3,0,0]
colors=["c","b","r","y","g"]
plt.pie(x,labels=classes,explode=explode,colors=colors,autopct="%0.1f%%")
plt.show()
```

```python
#To increase radius of circle
classes=["python","R","ML","AI","DS"]
x=[45,15,35,25,30]
explode=[0,0,0.3,0,0]
colors=["c","b","r","y","g"]
plt.pie(x,labels=classes,explode=explode,colors=colors,autopct="%0.1f%%",radius=1.4)
plt.show()
```

```python
classes=["python","R","ML","AI","DS"]
x=[45,15,35,25,30]
explode=[0,0,0.3,0,0]
colors=["c","b","r","y","g"]
plt.pie(x,labels=classes,explode=explode,colors=colors,autopct="%0.1f%%",startangle=270)
plt.show()
```

```python
#To enlarge the text size
classes=["python","R","ML","AI","DS"]
x=[45,15,35,25,30]
explode=[0,0,0.3,0,0]
colors=["c","b","r","y","g"]
textprops={"fontsize":15}
plt.pie(x,labels=classes,explode=explode,colors=colors,autopct="%0.1f%%",textprops=textprops)
plt.show()
```

```python
classes=["python","R","ML","AI","DS"]
x=[45,15,35,25,30]
explode=[0,0,0.3,0,0]
colors=["c","b","r","y","g"]
plt.pie(x,labels=classes,explode=explode,colors=colors,autopct="%0.1f%%")
plt.legend()
plt.show()
```
