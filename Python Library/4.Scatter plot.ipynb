```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x=np.random.randint(18,45,(100))
y=np.random.randint(15,440,(100))
print(x)
print(y)
```

```python
plt.scatter(x,y,marker="*",color="r")
```

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y,marker="*",color="r",linewidth="10",edgecolor="g")
plt.scatter(x,y,marker=".",color="y",linewidth="10",edgecolor="b")
plt.title("students attendence")
plt.xlabel("khan")
plt.ylabel("pathan")
plt.show
```
