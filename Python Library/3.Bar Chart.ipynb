```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
```

```python
classes=['python','R','AI','ML','DS']
x=[30,10,20,25,10]
y=[40,5,20,20,10]
z=[35,5,30,15,15]
```

```python
plt.bar(classes,x)
```

```python
plt.barh(classes,x)
```

```python
plt.bar(classes,x,width=0.2)
```

```python
plt.bar(classes,x,width=0.2,color='y')
```

```python
#To change the color of edge of graph
plt.bar(classes,x,width=0.2,color='y',edgecolor='R')
```

```python
plt.bar(classes,x,width=0.2,color='y',edgecolor='R',linewidth=5)
```

```python
plt.bar(classes,x,width=0.2,color='y',edgecolor='R',linewidth=5,alpha=1)
```

```python
#Line style
plt.bar(classes,x,width=0.2,color='y',edgecolor='R',linewidth=5,alpha=1,linestyle='--',label=x)
```

```python
#To hide the graph
plt.bar(classes,x,width=0.2,color='y',edgecolor='R',linewidth=5,alpha=1,linestyle='--',label=x,visible=False)
```

```python
#Merge the two graph
plt.figure(figsize=(16,9))
plt.bar(classes,x,width=0.2,color="b",label="class1 students")
plt.bar(classes,y,width=0.2,color="g",label="class2 students")
plt.title("Bar chart of students",fontsize=15)
plt.xlabel("classes",fontsize=15)
plt.ylabel("No. of students",fontsize=15)
plt.show()
```

```python
#To avoid overlapping of the graph
plt.figure(figsize=(16,9))
classes_index=np.arange(len(classes))
width=0.2
plt.bar(classes_index,x,width,color="b",label="class1 students")
plt.bar(classes_index+width,y,width,color="g",label="class2 students")
plt.title("Bar chart of students",fontsize=15)
plt.xlabel("classes",fontsize=15)
plt.ylabel("No. of students",fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(16,9))
classes_index=np.arange(len(classes))
width=0.2
plt.bar(classes_index,x,width,color="b",label="class1 students")
plt.bar(classes_index+width,y,width,color="g",label="class2 students")
plt.title("Bar chart of students",fontsize=15)
plt.xtricks(classes_index+width,classes)
plt.xlabel("classes",fontsize=15)
plt.ylabel("No. of students",fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(16,9))
classes_index=np.arange(len(classes))
width=0.2
plt.bar(classes_index,x,width,color="b",label="class1 students")
plt.bar(classes_index+width,y,width,color="g",label="class2 students")
plt.title("Bar chart of students",fontsize=15)
plt.xtricks(classes_index+width,classes,rotation=20)
```

```python
#To add another class
plt.figure(figsize=(16,9))
classes_index=np.arange(len(classes))
width=0.2
plt.bar(classes_index,x,width,color="b",label="class1 students")
plt.bar(classes_index+width,y,width,color="g",label="class2 students")
plt.bar(classes_index+width+width,z,width,color="y",label="class3 students")
```

```python
#To plot graph horizontally
plt.figure(figsize=(16,9))
classes_index=np.arange(len(classes))
width=0.2
plt.barh(classes_index,x,width,color="b",label="class1 students")
plt.barh(classes_index+width,y,width,color="g",label="class2 students")
plt.barh(classes_index+width+width,z,width,color="y",label="class3 students")
plt.ytricks(classes_index+width,classes,rotation=20)
plt.legend()
plt.show()
```
