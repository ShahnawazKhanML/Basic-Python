```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
z=sns.load_dataset("tips")
z
```

```python
sns.barplot(y=z.total_bill)
```

```python
#To visualize x and y axis
sns.barplot(x=z.day,y=z.total_bill)
```

```python
#For everytime we are writing z for simplisity(data=z)
sns.barplot(x="day",y="total_bill",hue="sex",data=z)
```

```python
#In the above graph order of day are randomly to make in a order(order)
A=['sun','Thur','Fri','sat']
sns.barplot(x="day",y="total_bill",hue="sex",data=z,order=A)
```

```python
#To interchange the order of bar plot(female is on first and male is on second)
B=["Female","Male"]
sns.barplot(x="day",y="total_bill",hue="sex",data=z,hue_order=B)
```

```python
7#To show the maximum range value on y axis(estimator)
#It is showing error to avoid it we have to import numpy
sns.barplot(x="day",y="total_bill",hue="sex",data=z,estimator=np.max)
```

```python
#To change the confidence intervel(CI)
sns.barplot(x="day",y="total_bill",hue="sex",data=z,ci=50)
```

```python
sns.barplot(x="day",y="total_bill",hue="sex",data=z,ci=20,n_boot=2)
```

```python
#To change the position of plot
sns.barplot(x="size",y="total_bill",hue="sex",data=z,ci=50,orient='h')
```

```python
sns.barplot(x="day",y="total_bill",hue="sex",data=z,color='g')
```

```python
#To change color randomly (palette)
sns.barplot(x="day",y="total_bill",hue="sex",data=z,palette='magma')
```

```python
sns.barplot(x="day",y="total_bill",hue="sex",data=z,palette='hot')
```

```python
sns.barplot(x="day",y="total_bill",hue="sex",data=z,saturation=0.3)
```

```python
sns.barplot(x="day",y="total_bill",hue="sex",data=z,saturation=20)
```

```python
sns.barplot(x="day",y="total_bill",hue="sex",data=z,saturation=10)
```

```python
#To change the color of CI Line
sns.barplot(x="day",y="total_bill",hue="sex",data=z,errcolor='0.5')
```

```python
sns.barplot(x="day",y="total_bill",hue="sex",data=z,errcolor='10')
```

```python
#To increase the size of size of CI
sns.barplot(x="day",y="total_bill",hue="sex",data=z,errwidth='10')
```

```python
#To change cap size
sns.barplot(x="day",y="total_bill",hue="sex",data=z,capsize=1)
```

```python
#To overlap the plot
sns.barplot(x="day",y="total_bill",hue="sex",data=z,dodge=False)
```

```python
#Uses of Kwargs
#(1)
kwargs={'alpha':0.9,'linestyle':':','linewidth':5,'edgecolor':'k'}
sns.barplot(x="day",y="total_bill",data=z,**kwargs)
```

```python
#(2)
sns.barplot(x="day",y="total_bill",data=z,alpha=0.9,linestyle="-",linewidth=6,edgecolor="g")
```

```python
#To perform the labelling,title ete
ax=sns.barplot(x="day",y="total_bill",data=z,alpha=0.9,linestyle="-",linewidth=6,edgecolor="g")
ax.set(title="Barplot of tips DataFrame",xlabel='Days',ylabel='Total_bill')
```

```python
#To See professionally
plt.figure(figsize=(16,9))
sns.barplot(x="day",y="total_bill",data=z,alpha=0.9,linestyle="-",linewidth=6,edgecolor="g")
plt.title=("Barplot of days & Totalbill",fontsize=20)
plt.xlabel=("Days",fontsize=20)
plt.ylabel=("Total_bill",fontsize=20)
plt.savefig("Barplot of Days & Total Bill")
```
