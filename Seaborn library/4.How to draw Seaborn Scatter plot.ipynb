```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
A=sns.load_dataset("titanic")
A
```

```python
sns.scatterplot(x="age",y="fare",data=A,hue="sex")
```

```python
sns.scatterplot(x="age",y="fare",data=A,hue="sex",style="who")
```

```python
#To visualize as per the age
sns.scatterplot(x="age",y="fare",data=A,hue="sex",style="who",size="who")
```

```python
#To zoomin the chart
sns.scatterplot(x="age",y="fare",data=A,hue="sex",style="who",size="who",sizes=(100,400))
```

```python
#To increase the size of graph
plt.figure(figsize=(16,9))
sns.scatterplot(x="age",y="fare",data=A,hue="sex",style="who",size="who",sizes=(100,400))
```

```python
#To see which is live in a bulk i.e man ,woman,child 
plt.figure(figsize=(16,9))
sns.scatterplot(x="age",y="fare",data=A,hue="alive",style="alive",size="who",sizes=(100,400))
```

```python
plt.figure(figsize=(16,9))
sns.scatterplot(x="age",y="fare",data=A)
sns.lineplot(x="age",y="fare",data=A)
sns.barplot(x="age",y="fare",data=A)
```
