```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
x=pd.read_csv("E:\\Data Science\ML by Indian AI\\4.Python Seaborn\\Who_is_responsible_for_global_warming.csv")
x
```

```python
x=x.drop(columns=["Country Code","Indicator Name","Indicator Code"],axis=1).set_index("Country Name")
x.head(6)
```
