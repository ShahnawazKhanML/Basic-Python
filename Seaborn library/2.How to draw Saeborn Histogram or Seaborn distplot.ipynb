```python
import seaborn as sns 
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
x=sns.load_dataset("tips")
x
```

```python
#To show histogram plot
#Fluctuating line is called as KDE(kernel density estimation)
sns.distplot(x["size"])
```

```python
sns.distplot(x["tip"],bins=100)
```

```python
#To see KDE only
sns.distplot(x["total_bill"],hist=False)
```

```python
#To see Histogram only
sns.distplot(x["total_bill"],kde=False)
```

```python
#To add another kde
#For this we have to import scipy library
sns.distplot(x["total_bill"],fit=norm)
```

```python
sns.distplot(x["total_bill"],kde=True,color="r")
```

```python
sns.distplot(x["total_bill"],kde=True,vertical=True)
```

```python
sns.distplot(x["total_bill"],kde=True,vertical=False)
```

```python
#To provide x axis name(axlabel)
sns.distplot(x["total_bill"],kde=True,axlabel="Total Bill")
```

```python
#To provide the label
#For this we have to import matplotlib
sns.distplot(x["total_bill"],kde=True,label="Bill Histogram")
plt.legend()
```

```python
#TO change the background color of plot and title for plot
plt.figure(figsize=(16,9))
sns.set()
sns.distplot(x["total_bill"],kde=True,label="Bill Histogram")
plt.title("Histogram of total bill")
plt.legend()
```

```python
x.total_bill.sort_values()
```

```python
#To divide the bar into the range(From 1 to 55) and show into bar
bins=[1,5,10,15,20,25,30,35,40,45,50,55]
plt.figure(figsize=(16,9))
sns.set()
sns.distplot(x["total_bill"],kde=True,label="Bill Histogram",bins=bins)
plt.xticks(bins)
plt.title("Histogram of total bill")
```

```python
#To change the histogram color
plt.figure(figsize=(16,9))
sns.set()
sns.distplot(x["total_bill"],kde=True,color="r")
```

```python
#To multiplre Histograms on single plot
#If you observe i selected that columns name which are in numeric data
plt.figure(figsize=(16,9))
sns.distplot(x["total_bill"],bins=9,label="total_bill")
sns.distplot(x["tip"],bins=9,label="tip")
sns.distplot(x["size"],bins=9,label="size")
plt.legend()
```
